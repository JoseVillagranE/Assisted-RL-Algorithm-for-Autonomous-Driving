```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npzfile = np.load("./../S_Rollouts/rollout_0_1.npz")
```

```python
npzfile
```

```python
states = npzfile['states']
actions = npzfile['actions']
rewards = npzfile["rewards"]
next_states = npzfile["next_states"]
terminals = npzfile["terminals"]
complementary_states = npzfile["complementary_states"]
next_complementary_states = npzfile["next_complementary_states"]
```

```python
print(f"states shape: {states.shape}")
print(f"actions shape: {actions.shape}")
print(f"rewards shape: {rewards.shape}")
print(f"next_states shape: {next_states.shape}")
print(f"terminals shape: {terminals.shape}")
print(f"complementary states shape: {complementary_states.shape}")
print(f"next complementary states shape: {next_complementary_states.shape}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
axes[0].plot(actions[:, 1]); axes[0].set_title("Throttle", fontsize=12, fontweight="bold")
axes[1].plot(actions[:, 0]); axes[1].set_title("Steer", fontsize=12, fontweight="bold")
fig.text(0.5, 0., 'Frame', ha='center', fontsize=16, fontweight="bold")
```
